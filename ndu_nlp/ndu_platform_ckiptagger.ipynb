載入必要模組

```python
from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
from collections import defaultdict
import gensim
```

輸入

```python
# 匯入停用詞
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords 

# 匯入預訓練檔案
ws = WS("./data")
pos = POS("./data")
ner = NER("./data")

# 匯入檔案，預設為testdata
documents = []
with open('./testdata.txt', 'r',encoding='utf-8') as file:
    documents = file.read().split("\n")

# 匯入停用詞
stopwords = stopwordslist('./中文停用詞.txt')  # 這裏加載停用詞的路徑

# 匯入辭典
with open('./自定義辭典(繁體).txt', 'r',encoding='utf-8') as file:
    user_dict = file.read().split("\n")
x = [i.split('\t') for i in user_dict]
word_to_weight = defaultdict()
for z in x :
    word_to_weight[z[0]] = z[1]
dictionary = construct_dictionary(word_to_weight)

# 確認是否要排除單詞還有進行雙詞判定
single_word = input('請輸入是否排除單詞: ')
n_gram = input('請輸入是否使用雙詞判定: ')
if n_gram == '是':
    n_gram_low = input('請輸入雙詞判定詞頻下限: ')
    n_gram_high = input('請輸入雙詞判定詞頻上限: ')
```

運算

```python
word_sentence_list = ws( documents, recommend_dictionary = dictionary )



# 斷詞
content = []
for sentence in word_sentence_list:
    sentence = ' '.join(sentence)
    result = ''
    words = sentence.split(' ')
    for word in words:
        if word not in stopwords:
            result = result + word + " "
    



# 多詞判定
if n_gram == '是':
    x = [i.split(" ") for i in content]
    bigram = gensim.models.Phrases(x, min_count=int(n_gram_low), threshold=int(n_gram_high)) 
    trigram = gensim.models.Phrases(bigram[x], min_count=int(n_gram_low), threshold=int(n_gram_high))  

    bigram_mod = gensim.models.phrases.Phraser(bigram)
    trigram_mod = gensim.models.phrases.Phraser(trigram)

    # See trigram example
    content = [' '.join(i) for i in trigram_mod[bigram_mod[x]]]
```
